```python
import pandas as pd

import matplotlib.pyplot as plt
import os
```

```python
def plot_number_collisions_step(N: int):
    path_to_folder = os.path.join("..", "simulation_particles", "data", f"N-{N}")
    filename = os.path.join(path_to_folder, "number_collisions_step.csv")

    with open(filename, "r") as f:
        df = pd.read_csv(f)

    time = df.iloc[:, 0]
    number_of_collisions = df.iloc[:, 1]

    fig, ax = plt.subplots()
    ax.set_title("Medida de pressão momentânea na esfera central (N = 5e4)")
    ax.set_xlabel("Tempo (s)")
    ax.set_ylabel("Número de partículas colidindo com a esfera")

    ax.scatter(time, number_of_collisions)
    plt.show()
```

```python
plot_number_collisions_step(50000)
```
