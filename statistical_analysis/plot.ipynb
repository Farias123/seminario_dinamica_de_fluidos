```python
import matplotlib.pyplot as plt
import numpy as np

from file_reading import data_Nx_N_
```

```python
Nx = 100
N_ = 3

data_dict = data_Nx_N_(Nx, N_)
```

```python
def plot_Nx_N_(data_dict: dict) -> None:
  steps_data = data_dict["steps_data"]

  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')

  indexes_simulated = np.unique(steps_data["idx"].values)
  for i in indexes_simulated:
    data_particle_i = steps_data[steps_data["idx"] == i]
    x_data = np.array(data_particle_i["x"])
    y_data = np.array(data_particle_i["y"])
    z_data = np.array(data_particle_i["z"])

    ax.plot(x_data, y_data, z_data, label=f"Particle {i}")

  ax.plot([0], [0], [0])

  # ax.legend()
  plt.show()

plot_Nx_N_(data_dict)
```
